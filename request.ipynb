```python
import json
import requests
from PIL import Image
import io
```

```python
url = 'http://127.0.0.1:8000/image_to_image'
```

```python
url = 'https://www.brisbane.qld.gov.au/sites/default/files/images/2020-12/200x177-keeping-a-dog.jpg'
response = requests.get(url)
img = Image.open(io.BytesIO(response.content))
img
```

```python
img_byte_arr = io.BytesIO()
img.save(img_byte_arr, format='PNG')
img_byte_arr.seek(0)
```

```python
img_byte_arr.getvalue()
```

```python
received = requests.post(url, data=img_byte_arr.getvalue())
received.content
```

```python
Image.open(io.BytesIO(received.content))
```
